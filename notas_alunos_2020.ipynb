**Agrupando as notas**

```python
agrupandoidnota = alunos_notas[['nr_anoletivo', 'cd_aluno', 'cd_turma', 'cd_periodo', 'ds_disciplina', 'ds_grupodisciplina', 'nr_nota']]
```

**Agrupando a nota de cada aluno individualmente por ano letivo, id do aluno, período, turma, grupo da disciplina e disciplina** 

Podemos ver a nota do aluno por período.

```python
mediaNotaPeriodo=agrupandoidnota.groupby(['nr_anoletivo', 'cd_aluno', 'cd_periodo', 'cd_turma', 'ds_grupodisciplina', 'ds_disciplina']).mean().reset_index()
```

```python
mediaNotaPeriodo
```

**Média anual de cada aluno em cada disciplina**

```python
mediaNotaAnual=agrupandoidnota.groupby(['nr_anoletivo', 'cd_aluno', 'cd_turma', 'ds_grupodisciplina', 'ds_disciplina']).mean().reset_index()
```

```python
mediaNotaAnual.drop('cd_periodo', axis=1, inplace=True)
```

```python
mediaNotaAnual
```

```python
mediaNotaAnual[mediaNotaAnual['nr_nota']<6]
```

**Fazendo a média total das disciplinas do aluno e criando a coluna resultados para conferir se o aluno foi aprovado ou reprovado**

Será atribuído 1 para aprovado, se o aluno ficou com nota média anual nas disciplinas >=6 e reprovado se o aluno ficou com nota média anual nas disciplinas <6.

```python
mediaNotaAnualTotal=agrupandoidnota.groupby(['nr_anoletivo', 'cd_aluno', 'cd_turma']).mean().reset_index()
```

```python
mediaNotaAnualTotal.drop('cd_periodo', axis=1, inplace=True)
```

**Notas médias anuais dos alunos => considerando a média de todas as disciplinas**

```python
mediaNotaAnualTotal
```

**Criando a coluna resultados**

```python
status =[]
for i in mediaNotaAnualTotal['nr_nota']:
  if i >= 6:
    status.append(1)
  else:
    status.append(0)
```

```python
status
```

```python
mediaNotaAnualTotal['Resultados'] = status
```

```python
mediaNotaAnualTotal
```

**Quantidade de reprovados**

```python
mediaNotaAnualTotal[mediaNotaAnualTotal['Resultados']==0].count()
```

**Quantidade de aprovados**

```python
mediaNotaAnualTotal[mediaNotaAnualTotal['Resultados']==1].count()
```

**Quantidade de alunos por turma**

```python
alunosportumra= mediaNotaAnualTotal[['cd_turma', 'Resultados']]
```

```python
alunosportumra.groupby('cd_turma').count()
```

**Quantidade de aprovados**

```python
aprovadotumra= mediaNotaAnualTotal[['cd_turma', 'Resultados']][mediaNotaAnualTotal['Resultados']==1]
```

```python
aprovadotumra.groupby('cd_turma').count()
```

```python
reprovadotumra= mediaNotaAnualTotal[['cd_turma', 'Resultados']][mediaNotaAnualTotal['Resultados']==0]
```

**Quantidade de reprovados**

```python
reprovadotumra.groupby('cd_turma').count()
```
